```python
file = open("D:/MDS/5196/Group026.txt", 'r')
#file = open("D:/MDS/5196/test2.txt", 'r')
line = file.read()
```

```python
import re
grants = re.finditer(r'<us-patent-grant.*>(.*\n)*?<\/us-patent-grant>', line)
#grants.group(0)
```

```python
grant_list = []
for i in grants:
    grant_list.append(i.group(0))
print(len(grant_list))
```

```python
print(grant_list[-1][-20:])
```

```python
grant_id = []
for i in grant_list:
    id = re.search(r'<us-patent-grant.*file="([\w]+)*-.*>', i)
    grant_id.append(id.group(1))
print(len(grant_id))
```

```python
title_list =[]
for i in grant_list:
    title = re.search(r'<invention-title id=.*>(.*)<\/invention-title>', i)
    title_list.append(title.group(1))
print(len(title_list))
```

```python
number_claims = []
for i in grant_list:
    claim = re.search(r'<number-of-claims>(\d{0,4})<\/number-of-claims>', i)
    if claim is not None:
        number_claims.append(claim.group(1))
    else:
        number_claims.append(0)
print(len(number_claims))
```

```python
name_list = []
for i in grant_list:
    name = re.search(r'<inventor.*>(.*\n)*?<last-name>(.*)<\/last-name>\n<first-name>(.*)<\/first-name>', i)
    name = name.group(3) + " " + name.group(2)
    name_list.append(name)
print(len(name_list))
#### some case multiple inventor -- neeeds more work :)
```

```python
cited_examiner = 0
cited_applicant = 0
cited_examiner_list = []
cited_applicant_list = []
for i in grant_list:
    citation = re.finditer(r'<us-citation>(.*\n)*<category>(.*)<\/category>', i)
    for c in citation:
        print(c.group(2))
        if c.group(2) == "cited by examiner":
            cited_examiner += 1
        elif c.group(2) == "cited by applicant":
            cited_applicant += 1
        else:
            continue
    cited_examiner_list.append(cited_examiner)
    cited_examiner = 0
    cited_applicant_list.append(cited_applicant)
    cited_applicant = 0
print(len(cited_applicant_list))
print(len(cited_examiner_list))
##### have to sub divide all citations as one string --> return 1 all the time
```
