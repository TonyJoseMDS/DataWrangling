```python
import re
tp_code=[]
ns_code=[]
bk_code=[]
outlier_tp=outlier.loc[outlier['branch_code'] == "TP"]
outlier_ns=outlier.loc[outlier['branch_code'] == "NS"]
outlier_bk=outlier.loc[outlier['branch_code'] == "BK"]


for row in outlier_tp['order_id'].iteritems():
    tpcode=re.findall(r'ORD(.)',row[1])
    for i in tpcode : 
            tp_code.append(i)
print(set(tp_code))

for row in outlier_ns['order_id'].iteritems():
    nscode=re.findall(r'ORD(.)',row[1])
    for i in nscode : 
            ns_code.append(i)
print(set(ns_code))

for row in outlier_bk['order_id'].iteritems():
    bkcode=re.findall(r'ORD(.)',row[1])
    for i in bkcode : 
            bk_code.append(i)
print(set(bk_code))
```

```python
for index, row in missing_data.iterrows():
    if str(row['branch_code']) == 'nan':
        code=re.search(r'ORD(.)',row['order_id']).group(1)
        code=code.strip()
        if code in set(tp_code):
            print('TP')
            missing_data.at[index,'branch_code']='TP'
            print(index,missing_data['branch_code'])
        elif code in set(ns_code):
            missing_data.at[index,'branch_code']='NS'
            #print(index,missing_data['branch_code'])
        elif code in set(bk_code):
            missing_data.at[index,'branch_code']='BK'
            #print(index,missing_data['branch_code'])


      
```

```python

for index, row in missing_data.iterrows():
    if str(row['distance_to_customer_KM']) == 'nan':
        #count+=1
        #print(row)
        source = (int(nodes.loc[(nodes['lat'] == row['customer_lat']) & (nodes['lon'] == row['customer_lon'])]['node']))
        if row['branch_code'] == 'BK':
            #print('BK')
            dist = (nx.dijkstra_path_length(network, source, bk, weight='length'))/1000
        elif row['branch_code'] == 'NS':
            #print('NS')
            dist = (nx.dijkstra_path_length(network, source, ns, weight='length'))/1000
        elif row['branch_code'] == 'TP':
            #print('TP')
            dist = (nx.dijkstra_path_length(network, source, tp, weight='length'))/1000
        missing_data.at[index, 'distance_to_customer_KM'] = dist
```

```python
missing_data['date'] = pd.to_datetime(missing_data['date'])
missing_data['day_of_week'] = missing_data['date'].dt.day_name()
def weekday_weekend(row):
    if row['day_of_week'] == 'Sunday':
        val = 1
    elif row['day_of_week'] == 'Saturday':
        val = 1
    else:
        val = 0
    return val

def time_of_day(row):
    if row['order_type'] == 'Breakfast':
        val = 0
    elif row['order_type'] == 'Lunch':
        val = 1
    elif row['order_type'] == 'Dinner':
        val = 2
    return val
```

```python
missing_data['weekend/weekday'] = missing_data.apply(weekday_weekend, axis=1)
missing_data['time_of_day'] = missing_data.apply(time_of_day, axis=1)

for index, row in missing_data.iterrows():
    if (row['customerHasloyalty?'] == 1):
        #print(row['delivery_fee'])
        #print(row['delivery_fee']*2)
        missing_data.at[index,'delivery_fee']=row['delivery_fee']*2
```

```python
#https://www.kdnuggets.com/2017/02/removing-outliers-standard-deviation-python.html 

import numpy

arr = [10, 386, 479, 627, 20, 523, 482, 483, 542, 699, 535, 617, 577, 471, 615, 583, 441, 562, 563, 527, 453, 530, 433, 541, 585, 704, 443, 569, 430, 637, 331, 511, 552, 496, 484, 566, 554, 472, 335, 440, 579, 341, 545, 615, 548, 604, 439, 556, 442, 461, 624, 611, 444, 578, 405, 487, 490, 496, 398, 512, 422, 455, 449, 432, 607, 679, 434, 597, 639, 565, 415, 486, 668, 414, 665, 763, 557, 304, 404, 454, 689, 610, 483, 441, 657, 590, 492, 476, 437, 483, 529, 363, 711, 543]

elements = numpy.array(arr)
print(elements)
mean = numpy.mean(elements, axis=0)
sd = numpy.std(elements, axis=0)

final_list = [x for x in arr if (x > mean - 3 * sd)]
final_list = [x for x in final_list if (x < mean + 3 * sd)]
print(len(final_list))
print(final_list)
```
