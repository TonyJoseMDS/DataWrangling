```python
nodes=pd.read_csv('nodes.csv')
nodes.head()
```

```python
import networkx as nx
```

```python
ed = pd.read_csv('edges.csv')
```

```python
bk = int(nodes.loc[(nodes['lat'] == -37.8158343) & (nodes['lon'] == 145.04645)]['node'])
ns = int(nodes.loc[(nodes['lat'] == -37.7738026)]['node'])
tp = int(nodes.loc[(nodes['lat'] == -37.8618349)]['node'])
print(ns, bk, tp)
#print(ns)
```

```python
network = nx.Graph()

network.add_nodes_from([ns, bk, tp])

for index, row in ed.iterrows():
    print(row['distance(m)'])
    network.add_edge(row['u'], row['v'], length = row['distance(m)'])
```

```python
#count = 0
for index, row in missing_data.iterrows():
#     print(row['branch_code'])
#     count +=1 
#     if count == 20:
#         break;
    if str(row['branch_code']) == 'nan':
        #count+=1
        #print(row['customer_lat'], "askjgdfka")
        source = (int(nodes.loc[(nodes['lat'] == row['customer_lat']) & (nodes['lon'] == row['customer_lon'])]['node']))
        #print()
        bk_dist = nx.dijkstra_path_length(network, source, bk, weight='length')
        ns_dist = nx.dijkstra_path_length(network, source, ns, weight='length')
        tp_dist = nx.dijkstra_path_length(network, source, tp, weight='length')
        #print(bk_dist, ns_dist, tp_dist)
        if (bk_dist < ns_dist) & (bk_dist < tp_dist):
            #print(bk_dist)
            branch = 'BK'
        elif (ns_dist < tp_dist) & (ns_dist < bk_dist):
            #print(ns_dist)
            branch = 'NS'
        else:
            #print(tp_dist)
            branch = 'TP'
        #print(branch)
        missing_data.at[index, 'branch_code'] = branch
```
